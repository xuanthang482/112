```
import os, cv2
from google.colab.patches import cv2_imshow
import imutils
image_file = {}
for file in os.listdir(data_path):
  image_file[file] = [os.path.join(data_path, file, img) for img in os.listdir(data_path+'/'+file)]
```

```
features_train, features_test, labels_train, labels_test = [], [], [], []
step = 0
leafID = 0
for name in image_file:
  index = 0
  for image in image_file[name]:
    print('{}/637'.format(step))
    step+=1
    leaf = cv2.imread(image)
    leaf = cv2.cvtColor(leaf, cv2.COLOR_BGR2GRAY)
    w = int(leaf.shape[1] / 10)
    leaf = imutils.resize(leaf, width=w)
    leaf = cv2.resize(leaf, (250, 440), interpolation=cv2.INTER_AREA)
    index+=1
    if index<=16:
      features_train.append(leaf)
      labels_train.append(leafID)
    else:
      features_test.append(leaf)
      labels_test.append(leafID)
  leafID+=1
```
